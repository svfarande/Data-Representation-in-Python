# Histograms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
fig, axs = plt.subplots(2,2,sharex=True)
axs = axs.flatten()
for i in range(len(axs)):
    sample_size = 10**(i+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[i].hist(sample) # default bins = 10
    axs[i].set_title(f'i={sample_size}')
```

```python
fig, axs = plt.subplots(2,2,sharex=True)
axs = axs.flatten()
for i in range(len(axs)):
    sample_size = 10**(i+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[i].hist(sample, bins=100)
    axs[i].set_title(f'i={sample_size}')
```

```python
plt.figure()
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
plt.scatter(X,Y)
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_hist = plt.subplot(gspec[0, 1:])
side_hist = plt.subplot(gspec[1:, 0])
lower_hist = plt.subplot(gspec[1:, 1:])
```

```python
top_hist.hist(X,bins=100)
s = side_hist.hist(Y, bins=100, orientation='horizontal')
lower_hist.scatter(X, Y)
```

```python
plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_hist = plt.subplot(gspec[0, 1:])
side_hist = plt.subplot(gspec[1:, 0])
lower_hist = plt.subplot(gspec[1:, 1:])

top_hist.clear()
top_hist.hist(X, bins=100, density=True)
side_hist.hist(Y, bins=100, orientation='horizontal', density=True)
side_hist.invert_xaxis()
lower_hist.scatter(X, Y)
```

```python
for ax in [top_hist, lower_hist]:
    ax.set_xlim(0,1)
for ax in [side_hist, lower_hist]:
    ax.set_ylim(-5,5)
```
