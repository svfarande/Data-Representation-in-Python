```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
linear_data = np.arange(1,10)
plt.figure()
plt.subplot(1,2,1)
plt.plot(linear_data, '-o')
```

```python
exponential_data  = linear_data ** 2
plt.subplot(1,2,2)
plt.plot(exponential_data,'-s')
```

```python
plt.subplot(1,2,1)
plt.plot(exponential_data, '-s', color='r')
plt.legend(['Linear', 'Exponential'])
```

```python
plt.figure()
ax1 = plt.subplot(1,2,1)
plt.plot(linear_data, '-o')
ax2 = plt.subplot(1,2,2, sharey=ax1)
plt.plot(exponential_data, '-s')
```

```python
fig, ax = plt.subplots(3,3, sharex=True, sharey=True)
ax[1,1].plot(linear_data, '-')
ax[2,1].plot(exponential_data, '-')
```

```python
# to display label for inner subplots
for iax in ax.flatten():
    iax.xaxis.set_tick_params(labelbottom=True)
    iax.yaxis.set_tick_params(labelbottom=True)
```

`plt.subplot()` is state-based & `plt.subplots()` is object based  
`plt.sublot()` uses plt for plotting (scripting layer)  
axes returned by `plt.subplots()` uses axes for plotting (artist layer)
