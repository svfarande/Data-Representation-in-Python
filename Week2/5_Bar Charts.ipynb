```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
linear_data = np.arange(1,9)
linear_data
```

```python
plt.figure()
xvals = range(1,len(linear_data)+1)
plt.bar(xvals,linear_data,width=0.3,color='blue')
```

```python
new_vals = []
for item in xvals:
    new_vals.append(item+0.3)
new_vals
```

```python
quadratic_data = linear_data ** 2
quadratic_data
```

```python
plt.bar(new_vals, quadratic_data, width=0.3, color='red')
```

```python
linear_err = np.array(np.random.randint(1,15,len(linear_data)))
linear_err
```

```python
plt.bar(xvals, linear_data, width=0.3, yerr=linear_err, color='blue')
plt.legend(['Linear', 'Quadratic'])
plt.title('Linear $x$ vs Quadratic $x^2$')
```

```python
plt.figure()
plt.bar(xvals, linear_data, width=0.3, color='b')
plt.bar(xvals, quadratic_data, width=0.3, color='g', bottom=linear_data)
plt.legend(['Linear', 'Quadratic'])
plt.title('Linear $x$ vs Quadratic $x^2$')
```

```python
plt.figure()
plt.barh(xvals, linear_data, height=0.3, color='b')
plt.barh(xvals, quadratic_data, height=0.3, color='g',left=linear_data)
plt.legend(['Linear', 'Quadratic'])
plt.title('Linear $x$ vs Quadratic $x^2$')
```

### Plot number of NULL value in dataframe

```python
df = pd.DataFrame(np.reshape(np.random.randint(10,99,8*8),(8,8)))
df
```

```python
df.iloc[5,2:6] = np.nan
df.iloc[0,3:8] = np.nan
df.iloc[3,5:6] = np.nan
df.iloc[2,7:8] = np.nan
df.iloc[4,1:8] = np.nan
df.iloc[7,4:7] = np.nan
df.columns = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight']
df
```

```python
df.isna().sum().plot(kind='bar')
```
