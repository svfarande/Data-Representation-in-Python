# Scatterplots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x = np.array(np.arange(1,9))
y = x
x
```

```python
plt.scatter(x,y)
```

```python
colours = ['green']*(len(x)-1)
colours.append('red')
plt.scatter(x,y,c=colours,s=100)
```

```python
zip_generator = zip(list(range(1,6)), list(range(6,11)))
x, y = zip(*zip_generator)
x
```

```python
y
```

```python
plt.scatter(x[:2],y[:2], s=50, c='red', label='Short students')
```

```python
plt.scatter(x[2:],y[2:], s=50, c='green', label='Tall students')
```

```python
plt.xlabel('The number of times the child kicked a ball')
plt.ylabel('The grade of student')
plt.title('Relationship between ball kicking and grades')
```

```python
plt.legend() # Default top left corner
```

```python
plt.legend(loc=4, frameon=False, title='Legend') # 4 is lower right corner
```

```python
plt.gca().get_children()
```

```python
legend = plt.gca().get_children()[-2]
legend.get_children()[0].get_children()[1].get_children()[0].get_children()
```

```python
from matplotlib.artist import Artist

def rec_gc(art, depth=0):
    if isinstance(art, Artist):
        print(" " * depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)

rec_gc(legend)
```
