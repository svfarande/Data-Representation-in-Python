```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib notebook
```

```python
linear_data = np.array(np.arange(1,9))
linear_data
```

```python
quadratic_data = linear_data**2
quadratic_data
```

```python
plt.plot(linear_data, '-o', quadratic_data, '-o')
```

```python
plt.plot([11,22,33,44,55,66], '--r')
```

```python
plt.xlabel('X-axis (Index)')
plt.ylabel('Y-axis (Values)')
plt.title('A title')
plt.legend(['linear','quadratic','*11'])
```

```python
# highlighting difference between orange and blue curve
plt.gca().fill_between(range(len(linear_data)), linear_data, quadratic_data, facecolor='blue',
                      alpha=0.25)
```

```python
observation_dates  = np.arange('2017-01-01','2017-01-09',dtype='datetime64[D]')
# observation_dates = list(map(pd.to_datetime, observation_dates))
observation_dates
```

```python
plt.figure()
plt.plot(observation_dates,linear_data,'-o',
         observation_dates,quadratic_data,'-o')

plt.xticks(rotation=45)
'''
x = plt.gca().xaxis
for item in x.get_ticklables():
    item.set_rotation(45)
'''

plt.subplots_adjust(bottom=0.25)

ax=plt.gca()
ax.set_xlabel('Dates')
ax.set_ylabel('Values')
ax.set_title('Quadratic($x^2$) vs Linear($x$)')

plt.legend(['Linear','Quadratic'])
```
